```python
import matplotlib.pyplot as plt
import pandas

data = pandas.DataFrame
with open("runtime.csv") as csvfile:
    data = pandas.read_csv(csvfile)
    data["holetot"] = data["adj"]+data["holef"]
    data["minktot"] = data["adj"]+data["mink"]
    data["Method 1"] = data["holetot"]/60
    data["Method 2"] = data["minktot"]/60

ax=plt.gca()
data.plot(kind="line", x="nobuilds",y="Method 1",color="red", ax=ax)
data.plot(kind="line", x="nobuilds",y="Method 2",color="green",ax=ax)
plt.show()
```
